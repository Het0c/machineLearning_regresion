```python
# Importaciones
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
df = pd.read_csv(name, sep=",")
df.head(10)
```

```python
# Filtrar los primeros 151 Pokémon
df_filtrado = df.head(151)


# Crear el gráfico de dispersión
plt.figure(figsize=(10, 8))  # Ajustar el tamaño del gráfico (opcional)
sns.scatterplot(x='Attack', y='Sp. Atk', hue= , data=df_filtrado)
plt.title('Relación entre Ataque y Ataque Especial de Pokémon')
plt.xlabel('Attack')
plt.ylabel('Sp. Atk')
plt.show()
```
